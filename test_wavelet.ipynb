```python
import pywt

x = np.random.randint(10,size=1*500)

cA, cD = pywt.dwt(data=x, wavelet='db3')
```

```python
w = pywt.Wavelet('db3')
```

```python
params_w_maxlev = pywt.dwt_max_level(len(x),w)
params_w_maxlev
```

```python
pywt.downcoef('a', x, w, level=params_w_maxlev)
```

```python
print(cA)

print(cD)
```

```python
pywt.idwt(cA=cA,cD=cD,wavelet='db3')
```
